<h2>Data Cleaning </h2>

```python
df.columns
```

```python
##Remove multiple columns
df=df.drop(columns=['Poster_Link','Certificate','Overview'])
```

```python
#Extract only needed columns
df=df[['Series_Title','Released_Year','Genre','IMDB_Rating','No_of_Votes','Director']]
df.head()
```

```python
##Calculate total number of missing values in each column
df.isna().sum()
```

```python
##Calculate total number of duplicated rows in dataframe
df.duplicated().sum()
```

```python
##If duplicates occured
##df.drop_duplicates(inplace=True)
```

```python
df['Genre'] = df['Genre'].str.split(', ')
df.head()
```

<h2>Analysis </h2>

```python
##Function to categorize according to IMDB_Rating
def categorize_rating(rating):
    if rating >= 9:
        return "Masterpiece"
    elif 7 <= rating < 9:
        return "Very Good"
    elif 5 <= rating < 7:
        return "Mediocre"
    else:
        return "Poor"
```

```python
##Create a new column for Category of Rating whose row values will be returned from function
df['Rating_Category'] = df['IMDB_Rating'].apply(categorize_rating)
```

```python
df.head()
```

```python
df.tail()
```

```python
# Filter movies with more than 1000 votes
popular_movies = df[df['No_of_Votes'] > 1000]
```

```python
popular_movies
```

```python
# Sort by rating
top10 = popular_movies.sort_values(by='IMDB_Rating', ascending=False).head(10)
```

```python
print("\nTop 10 Highest Rated Movies:")
print(top10[['Series_Title', 'IMDB_Rating', 'No_of_Votes']])
```

```python
##Average Rating Per Genre
df_exploded = df.explode('Genre')
df_exploded
```

```python
avg_rating = df_exploded.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False)
print("\nAverage Rating per Genre:")
print(avg_rating)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=avg_rating.index, y=avg_rating.values, palette="viridis")
plt.xticks(rotation=45)
plt.ylabel("Average IMDB Rating")
plt.title("Average IMDB Rating per Genre")
plt.show()
```

```python
##Count Number of Movies Released per Year
movies_per_year = df['Released_Year'].value_counts().sort_index()
movies_per_year
```

```python
plt.figure(figsize=(12,6))
plt.plot(movies_per_year.index, movies_per_year.values, marker="o")
plt.title("Movies Released Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")

# Rotate labels
plt.xticks(rotation=45)

# Show only every 5th year
plt.xticks(movies_per_year.index[::5])
plt.show()
```

```python
##Avergae Rating Per Year
avg_rating_year = df.groupby('Released_Year')['IMDB_Rating'].mean()
avg_rating_year
```

```python
plt.figure(figsize=(12,6))
plt.plot(avg_rating_year.index, avg_rating_year.values, marker="o")
plt.title("Average Ratings by Year Year")
plt.xlabel("Year")
plt.ylabel("Average Ratings")

# Rotate labels
plt.xticks(rotation=45)

# Show only every 5th year
plt.xticks(avg_rating_year.index[::5])
plt.show()
```

```python
##Average Rating Per Director
avg_rating = df.groupby("Director")["IMDB_Rating"].mean()
avg_rating
```

```python
##Movie Count
movie_count = df.groupby("Director")["Series_Title"].count()
movie_count
```

```python
directors = pd.DataFrame({
    "avg_rating": avg_rating,
    "movie_count": movie_count
}).reset_index()
directors = directors[directors["movie_count"] >= 3]
directors
```

```python
top_directors = directors.sort_values("avg_rating", ascending=False).head(10)
print(top_directors)
```
